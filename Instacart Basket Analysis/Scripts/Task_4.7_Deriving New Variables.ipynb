# 2. If-Statements with User-Defined Functions

```python
# Define a function to sort price ranges/price_label
def price_label(row):
  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna=False)
```

```python
#Check the most expensive product within the subset
df['prices'].max()
```

# 3. If-Statements with the loc() Function

```python
#Create conditions
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna=False)
```

# 4. If-Statements with For-Loops

```python
# Check for the frequnecy of orders on each day of the week
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

The value 0 means Saturday, it has the highest frequency, which means that Saturday is the busiest day. 
The 4 value means Wednesday, it has the lowest frequency, which means the slowest day of Instacart app orders.

```python
# Define new variable 'busiest day' using for-loop
result = []
for value in df_ords_prods_merged['orders_day_of_week']:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
#Combine result with the dataframe (creating a new column and setting it equal to result)
df_ords_prods_merged ['busiest_day'] = result
```

```python
df_ords_prods_merged ['busiest_day'].value_counts(dropna=False)
```

# 5. Task 4.7

```python
#Create new variable 'busiest days' (two slowest and two busiest days)
result2 = []

for value in df_ords_prods_merged['orders_day_of_week']:
  if value in [0,1]:
    result2.append("Busiest days")
  elif value in [3,4]:
    result2.append("Slowest days")
  else:
    result2.append("Regularly busy")
```

```python
#Combine result with the dataframe (creating a new column and setting it equal to result2)
df_ords_prods_merged ['busiest_days'] = result2
```

```python
#Determine periods of time labeled Most orders, Average orders, and Fewest orders 

result3 = []

for value in df_ords_prods_merged['order_hour_of_day']:
  if value>=9 and value<=16:
    result3.append("Most orders")
  elif value<=6:
    result3.append("Fewest orders")
  else:
    result3.append("Average orders")
```

```python
#Combine the result (the time periods) with the dataframe ords_prods_merged 
df_ords_prods_merged['busiest_period_of_day'] = result3
df_ords_prods_merged['busiest_period_of_day'].value_counts()
```
